```python
import cv2
import numpy as np

# Load the images
img1 = cv2.imread('image1.jpg')
img2 = cv2.imread('image2.jpg')
img3 = cv2.imread('image3.jpg')

# Resize the images to the same dimensions
resized_img1 = cv2.resize(img1, (970, 1296))
resized_img2 = cv2.resize(img2, (970, 1296))
resized_img3 = cv2.resize(img3, (970, 1296))

# Create an array of the resized images
images = np.array([resized_img1, resized_img2, resized_img3])

# Calculate the average image
avg_img = np.mean(images, axis=0)

# Display the average image
cv2.imshow('Average Image', avg_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
